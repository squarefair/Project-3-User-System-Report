```python
filename = "userlog.log"
def suspiciousActivities(filename):
    infile = open(filename,"r")
    results = []
    users = dict()
    suspiciousUsers=[]
    lines = infile.readlines()
    lines.sort() 
    currentDate=int(lines[0].split("    \t\t")[0].split(" ")[0].split("-")[-1])
    words = lines[0].split("    \t\t")
    
    for line in lines:
        
        words = line.split("    \t\t")
        if(len(words)<4):
            continue
        time = words[0]
        activity = words[1]
        server = words[2]
        email = words[3]
        if currentDate != int(time.split(" ")[0].split("-")[-1]):
            suspiciousUsers=[]
            currentDate=int(time.split(" ")[0].split("-")[-1])
            users=dict()
            
        if currentDate == int(time.split(" ")[0].split("-")[-1]):
            if email in suspiciousUsers:
                continue
            
            if 0<=int(time.split(" ")[1].split(":")[1]) and int(time.split(" ")[1].split(":")[1])<=5:
                results.append(line)
                suspiciousUsers.append(email)
                if email in users.keys():
                  users[email]+=1
                else:
                    users[email]=1
                
            elif email in users.keys():
                users[email]+=1
                if users[email]==5:
                    results.append(line)
                    suspiciousUsers.append(email)
                else:
                    users[email]=1
    
    f = open("suspicious_report.txt", "w")
    f.write("Total Count of Suspicious Activities: "+str(len(results)))
    for result in results:
        f.write(result+"\n")
    f.close()
    
suspiciousActivities(filename)
```

```python
def irresponsibleBehavior (filename):
    infile = open(filename,"r")
    results = []
    users = dict()
    suspiciousUsers=[]
    lines = infile.readlines()
    lines.sort() 
    currentDate=int(lines[0].split("    \t\t")[0].split(" ")[0].split("-")[-1])
    words = lines[0].split("    \t\t")
   
    for line in lines:
        
        words = line.split("    \t\t")
        if(len(words)<4):
            continue
        time = words[0]
        activity = words[1]
        server = words[2]
        email = words[3]
        iUsers=[]
        if currentDate != int(time.split(" ")[0].split("-")[-1]):
            iUsers=[]
            currentDate=int(time.split(" ")[0].split("-")[-1])
            users=dict()
           
        if currentDate == int(time.split(" ")[0].split("-")[-1]):
            if email in iUsers:
                continue

            if email in users.keys():
                if activity == "login":
                    users[email][0]+=1
                else:
                    users[email][1]+=1
            else:
                users[email]=[0,0]
                if activity == "login":
                    users[email][0]+=1
                else:
                    users[email][1]+=1
            if users[email][0]+1 == users[email][1]:
                results.append(line)
                iUsers.append(email)
            
    
    f = open("irresponsible_report.txt", "w")
    f.write("Total Count of Irresponsible Activities: "+str(len(results)))
      
    for result in results:
        f.write(result+"\n")
    f.close()
    
irresponsibleBehavior(filename)
```

```python
def domainCount (filename):
    infile = open(filename,"r")
    results = []
    domains = dict()
    suspiciousUsers=[]
    lines = infile.readlines()
    lines.sort() 
    currentDate=int(lines[0].split("    \t\t")[0].split(" ")[0].split("-")[-1])
    words = lines[0].split("    \t\t")
 
    for line in lines:
        
          
        words = line.split("    \t\t")
        if(len(words)<4):
            continue
        time = words[0]
        activity = words[1]
        server = words[2]
        email,domain = words[3].split("@")
        
        if domain in domains.keys():
                if email in domains[domain]:
                    domains[domain][email]+=1
                else:
                    domains[domain][email]=1
                    
        else:
                domains[domain] = dict()
                if email in domains[domain]:
                    domains[domain][email]+=1
                else:
                    domains[domain][email]=1

    f = open("domain_report.txt", "w")
    f.write("Domain\tUser Count")
    for domain in domains.keys():
        f.write(domain+"\t"+str(len(domains[domain]))+"\n")
    
    f.close()
    
domainCount(filename)
```
